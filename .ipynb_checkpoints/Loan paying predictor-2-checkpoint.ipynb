```python
total_acc_avg = loan.groupby('total_acc').mean()['mort_acc']
```

```python
#Creating function to fill in these values
def mort_acc(total_acc,mort_acc):
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
loan['mort_acc'] = loan.apply(lambda x: mort_acc(x['total_acc'], x['mort_acc']), axis=1)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=new_loan.drop('paid',axis=1).values
y=new_loan['paid'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Model Creation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python
X_train.shape
```

```python
X_test.shape
```

### Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
```

### Dropout layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
#input layer
model.add(Dense(units=80,activation='relu'))
model.add(Dropout(0.5))
#Hidden layers
model.add(Dense(units=54,activation='relu')) 
model.add(Dropout(0.5))
#ouput layer
model.add(Dense(units=1,activation='sigmoid')) 
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=50,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

### Model Evaluation

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
